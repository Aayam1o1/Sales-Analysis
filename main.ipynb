# Data Preperation

Write a python program to merge data from each month into one CSV and read in updated datafram

```python
#importing pandas
import pandas as pd

#importing matplotlib
from matplotlib import pyplot as plt

import seaborn as sb
```

```python
#reading all the csv files and storing it in a dataframe
CSV_file_1 = pd.read_csv("CSV/Sales_January_2019.csv")
CSV_file_2 = pd.read_csv("CSV/Sales_February_2019.csv")
CSV_file_3 = pd.read_csv("CSV/Sales_March_2019.csv")
CSV_file_4 = pd.read_csv("CSV/Sales_April_2019.csv")
CSV_file_5 = pd.read_csv("CSV/Sales_May_2019.csv")
CSV_file_6 = pd.read_csv("CSV/Sales_June_2019.csv")
CSV_file_7 = pd.read_csv("CSV/Sales_July_2019.csv")
CSV_file_8 = pd.read_csv("CSV/Sales_August_2019.csv")
CSV_file_9 = pd.read_csv("CSV/Sales_September_2019.csv")
CSV_file_10 = pd.read_csv("CSV/Sales_October_2019.csv")
CSV_file_11 = pd.read_csv("CSV/Sales_November_2019.csv")
CSV_file_12 = pd.read_csv("CSV/Sales_December_2019.csv")

#Merging all CSV files into one merged CSV file using concat
Merged_CSV_file = pd.concat([CSV_file_1,CSV_file_2,CSV_file_3,CSV_file_4,CSV_file_5,CSV_file_6,CSV_file_7,
                             CSV_file_8,CSV_file_9,CSV_file_10,CSV_file_11,CSV_file_12])

#Storing merged data in one single csv file
Merged_CSV_file.to_csv ("merged_data.csv")


```

```python
#reading merged csv file and storing in dataframe 
Dataframe = pd.read_csv("merged_data.csv")
Dataframe
```

#  Data Exploration
Which Month has the best sales? and how much was the earning in that
month? Make a bar graph of sales as well. 

```python
#Since there are multiple dats with same month name we are usinmg unique keyword to get only one month name
unique_months = pd.unique(Dataframe['Month_name'])
unique_months.sort()

unique_months
```

```python
# Creating new column named total sum in dataframe and inseting total sum acquired by multiplying quantity and price
Dataframe['Total_sum'] = Dataframe['Quantity Ordered'] * Dataframe['Price Each'] 

# CREATING a variable nameed Total_sum_of_month and storing the total sum of month acquired by grouping monthname with
# total sum
Total_sum_of_month = Dataframe.groupby('Month_name')['Total_sum'].sum()
```

```python
#Creating new Dataframe named frame
frame = pd.DataFrame()

#Adding duplicate_months adn Total_sum_of_month.values into the dataframe "frame"
frame['Month_name']= unique_months
frame['sum'] = Total_sum_of_month.values

frame[['Month_name', 'sum']]
```

```python
#Finding the month with highest sales
frame[Total_sum_of_month.index == Total_sum_of_month.idxmax()]
```

```python
#Finding the month with lowest sales
frame[Total_sum_of_month.index == Total_sum_of_month.idxmin()]
```

```python
#Creating a bar graph with different colors
fig= plt.figure()
axes = fig.add_axes([0.1,0.1,1.5,1])
color = ['#D7B4DF', '#D8F1C2', '#74B9D3', '#FCA789', '#F6F7B6', '#c5e4dc', '#F7B6D7','#E39E4E', '#7E9997', '#7DCEA0',
         '#85929E', '#CCCCFF']
plt.ticklabel_format(style='plain')

# Adding the columns into the bar graph and giving labels
axes.bar(unique_months,Total_sum_of_month, width=0.8, color = color)

#Naming labels and setting title
axes.set_xlabel("Month")
axes.set_ylabel("Total Sales")
axes.set_title("Sales of each Month")

plt.show()
```
